```python
from wiktionaryparser import WiktionaryParser

parser = WiktionaryParser()
```

```python
from pprint import pprint
```

```python
word = 'bat'
response = parser.fetch(word)
pprint(response)
```

```python
import re

Dict = {}
count = 0
for meaning in response:
    count += 1
    word_class = ''
    word_def = []
    examples = []
    
    # filtering pronunciations.
    IPAs = {}
    for p in meaning['pronunciations']['text']:
        
        if 'IPA:' in p:
            dialects = ''
            if re.search('US|UK|Received Pronunciation|American|England|enPR|GenAm',p):
                UK = re.search('England|UK|Received Pronunciation', p)
                US = re.search('US|American|GenAm|enPR', p)
                if UK and 'UK' not in IPAs:
                    IPAs['UK'] = re.search('(IPA: )(.*)',p).group(2)
                if US and 'US' not in IPAs:
                    IPAs['US'] = re.search('(IPA: )(.*)',p).group(2)
            if re.match('IPA:',p) and 'IPA' not in IPAs: 
                IPAs['IPA'] = re.search('(IPA: )(.*)',p).group(2)
    if len(IPAs) == 3:
        del IPAs['IPA']
    if len(IPAs) == 2:
        if IPAs['UK'] == IPAs['US']:
            IPAs['IPA'] = IPAs['US']  
            del IPAs['UK'], IPAs['US']
            
            

    for definition in meaning['definitions']:
        word_class = definition['partOfSpeech']
        def_list = definition['text'] 
       
    #for df in definition['text']:
        #eliminate obsolet ones
        
        
        for Eg in definition['examples']:
            if len(Eg) < 40 and not re.match('Audio',p):
                examples.append(Eg)
       
        if len(def_list) == 0: continue
            
        data = {}    
        data['class'] = word_class
        data['def_list'] = def_list
        data['examples'] = examples
            
        if word in Dict:       
            if Dict[word]['pronunciation'] != IPAs:
                data['pronunciation'] = IPAs
            Dict[word]['meanings'].append(data) 
        else: 
            Dict[word] = {}
            Dict[word]['pronunciation']= IPAs
            Dict[word]['meanings'] = []
            Dict[word]['meanings'].append(data) 
    
# (UK, slang, obsolete)   (obsolete or dialectal)(Caribbean, MLE)

pprint(Dict) 





  # 
  #             
# return (word: {pronuntiation:'', meanings:[{class:'',def_list:[str],examples:[str]]})   
```
