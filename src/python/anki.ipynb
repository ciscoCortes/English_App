```python
import genanki
```

```python
model_id = 1
my_model = genanki.Model(
  model_id, 
  'Simple',
  fields=[
    {'name': 'Question'},
    {'name': 'Answer'}
    # {'name' : 'Sentences'}
  ],
  templates=[
    {
      'name': 'Card 1',
      'qfmt': '{{Question}}',
      'afmt': '{{Answer}}'
    }
  ],
    css = style)
```

```python
deck_id = 11
my_deck = genanki.Deck(deck_id = deck_id, 
                      name = 'Firts Level', 
                      description = 'English first level dictionary')
```

```python
import json
with open("ownDict.json") as f:
    data = json.loads(f.read())
```

```python
import re

question = ''
answer = ''
# bucle = 0
htmlA = ''
htmlQ = ''
egPos = ''
ejPos = ''
egWord = ''
ejWord = ''
for word in data.keys(): 
    # if bucle == 4: break
    # bucle += 1
    morfo = data[word]["morfology"]
    meaningsEg = ""
    meaningsEj = ""
    ipa = ""
    pattern = rf"\b{word.capitalize()}\b|\b{word}\b"
    if "US" in data[word]["pronunciation"]:
        ipa = data[word]["pronunciation"]['US']
    elif "IPA" in data[word]["pronunciation"]:
        ipa = data[word]["pronunciation"]['IPA']
    for meaning in data[word]["meanings"]:
        wordClass = meaning["class"]
        count = 0
        examples = ""
        examples_trans = ""
        for eg in meaning["examples"]:
            if len(meaning['ejemplos']) > count:
                if count > 1:
                    break
                eg = re.sub(rf'\b{word}\b', '<span class="egWord">{}</span>'.format(word),eg)    
                eg = re.sub(rf'\b{word.capitalize()}\b', '<span class="egWord">{}</span>'.format(word.capitalize()),eg)    
                ej = meaning['ejemplos'][count]
                trans = meaning["trans_terms"][count]
                ej = re.sub(rf'\b{trans}\b', '<strong>{}</strong>'.format(trans),ej)    
                ej = re.sub(rf'\b{trans.capitalize()}\b', '<strong>{}</strong>'.format(trans.capitalize()),ej)    
                example = "<p class='example'>{}</p>".format(eg)
                example_ej = example + "<p class='ejemplo'>{}</p>".format(ej)
                examples = examples + example
                examples_trans = examples_trans + example_ej
                count +=1
        meaningEg = """
            <fieldset>
                <legend>{}</legend>
                {}
            </fieldset>
        """.format(wordClass, examples)
        meaningEj = """
            <fieldset>
                <legend>{}</legend>
                {}
            </fieldset>
        """.format(wordClass, examples_trans)
            
        meaningsEg = meaningsEg + meaningEg
        meaningsEj = meaningsEj + meaningEj
    htmlQ = """
        <div class="title">
        <strong class="word">{}</strong>
        <span class="IPA">{}</span>
        <p class="morfo">{}</p>
        </div>
        <br>
        <main>
        {}
        </main>
    """.format(word, ipa, morfo, meaningsEg)
    htmlA = """
        <div class="title">
        <strong class="word">{}</strong>
        <span class="IPA">{}</span>
        <p class="morfo">{}</p>
        </div>
        <br>
        <main>
        {}
        </main>
    """.format(word, ipa, morfo, meaningsEj)

    card = genanki.Note(model = my_model, 
                       fields = [
                                htmlQ, 
                                htmlA
                                ])
    my_deck.add_note(card)
```

```python
genanki.Package(my_deck).write_to_file('First_Level.apkg')
```
